```python
import pandas as pd
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_location = "./melb_data.csv"
data = pd.read_csv(data_location)
data
```

```python
print("Descripicon estadistica: ")
print(data.describe())
```

```python
print("Informacion dataset: ")
print(data.info())
```

```python
print("Conteo valores unicos")
print(data['Rooms'].value_counts())
```

```python
print("Valores nulos")
print(data.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data["Rooms"],kde=False)
plt.show
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data["Rooms"],shade=True)
plt.show
```

Analsisi de correalacion

```python
select_columns = ["Rooms", "Bathroom", "Bedroom2", "Landsize"]
data2 = data[select_columns]
data2
```

```python
print("Matriz de correlacion")
correlacion_matriz = data2.corr()
print(correlacion_matriz)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlacion_matriz,annot=True,cmap='coolwarm')
plt.title("Heatmap")
plt.show
```
